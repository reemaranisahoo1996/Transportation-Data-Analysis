#IMPORTING LIBRARIES

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
```

#Loading CSV File

```python
data = pd.read_csv("/content/Assignment (Individual)_1.csv")
print(data.head(15))
```

```python

data.drop(data.columns[13:51], axis=1, inplace=True, errors='coerce')
data.drop(data.index[70:590], axis=0, inplace=True, errors="coerce")
data.columns = ['Age',
              'Gender',
              'Education',
              'Income',
              "Occupation",
              "Priv_Vehicle",
              "HH_Size",
              "Travel Time",
              "Type",
              "Scenario_1",
              "Scenario_2",
              "Scenario_3",
              "Scenario_4"]

print("Original NaN Value")
print(data[['Age',
          'Gender',
          'Education',
          'Income',
          "Occupation",
          "Priv_Vehicle",
          "HH_Size",
          "Travel Time",
          "Type",
          "Scenario_1",
          "Scenario_2",
          "Scenario_3",
          "Scenario_4"]].isnull().sum())

print(data.dtypes)

```

# Conversion of object datatype to numeric datatype

```python

data = data.replace(to_replace={'Female': 0, 'Male': 1})

data = data.replace(to_replace={"secondary": 0,
                            "higher secondary": 1,
                            "graduate & above": 2,
                            })

data = data.replace(to_replace={"<6,000": 0,
                            "6,000- 10,000": 1,
                            "10,000- 20,000": 2,
                            "20,000- 50,000": 3,
                            "2,00,000- 5,00,000": 4,
                            ">5,00,000": 5,
                            })

data = data.replace(to_replace={"unemployed": 0,
                            'Student': 1,
                            "private employee": 2,
                            "govt. employee": 3,
                            "employed for wages": 4,
                            "self employed": 5
                            })

data = data.replace(to_replace={'Yes': 1, 'No': 0})
data = data.replace(to_replace={'type 1': 1})
```

#DATA FILLING

```python

data["Gender"].fillna(data["Gender"].mean(), inplace=True)
data["Education"].fillna(data["Education"].mean(), inplace=True)
data["Income"].fillna(data["Income"].mean(), inplace=True)
data["Occupation"].fillna(data["Occupation"].mean(), inplace=True)
data["Priv_Vehicle"].fillna(data["Priv_Vehicle"].mean(), inplace=True)
data["HH_Size"].fillna(data["HH_Size"].mean(), inplace=True)
data["Travel Time"].fillna(data["Travel Time"].mean(), inplace=True)
```

```python
data.head(15)
```

```python
data["Scenario_1"].fillna(data["Scenario_1"].mean(), inplace=True)
data["Scenario_1"] = data["Scenario_1"].round(decimals=0)
data["Scenario_2"].fillna(data["Scenario_2"].mean(), inplace=True)
data["Scenario_2"] = data["Scenario_2"].round(decimals=0)
data["Scenario_3"].fillna(data["Scenario_3"].mean(), inplace=True)
data["Scenario_3"] = data["Scenario_3"].round(decimals=0)
data["Scenario_4"].fillna(data["Scenario_4"].mean(), inplace=True)
data["Scenario_4"] = data["Scenario_4"].round(decimals=0)

data['Predict'] = data.loc[:, ["Scenario_1", "Scenario_2", "Scenario_3", "Scenario_4"]].sum(axis=1)

scale = MinMaxScaler()
data[["Predict"]] = scale.fit_transform(data[["Predict"]])
data["Predict"] = data["Predict"].round(decimals=0)
print(data.head(15))
```

# IMPORTING PACKAGES FROM SKLEARN LIBRARY

```python

from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
```

```python

feature_cols = ['Age',
                'Gender',
                'Education',
                'Income',
                "Occupation",
                "Priv_Vehicle",
                "HH_Size",
                "Travel Time",
                "Type",
                "Scenario_1",
                "Scenario_2",
                "Scenario_3",
                "Scenario_4"]

X = data[feature_cols] 
Y = data.Predict
```

# MODEL TRAINING/TESTING

```python

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

modeling = tree.DecisionTreeClassifier()

modeling1= modeling.fit(X_train, y_train)

y_predicted = modeling1.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_predicted))
print("Precision: ")
```

# precision = TP/TP+FP
# recall = TP/TP+FN

```python


micro_precision = precision_score(y_predicted, y_test, average='micro')
print('Micro-averaged precision score: {0:0.2f}'.format(
    micro_precision))

macro_precision = precision_score(y_predicted, y_test, average='macro')
print('Macro-averaged precision score: {0:0.2f}'.format(
    macro_precision))

class_precision = precision_score(y_predicted, y_test, average=None)
print('Per-class precision score:', class_precision)

recall = recall_score(y_test, y_predicted, average=None)
print("Recall scrore: ", recall)

cnf_matrix = confusion_matrix(y_test, y_predicted)
print('Confusion matrix: \n', cnf_matrix)

clas_report = classification_report(y_test, y_predicted)
print("Classification report: \n", clas_report)
```
